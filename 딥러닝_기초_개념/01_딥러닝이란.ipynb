# 인공신경망

## 1. 인공신경망 정의 (Artificial Neural Network)
- **인간 두뇌에 대한 계산적 모델**을 통해 인공지능을 구현하려는 분야
- 인간의 뇌 구조 모방 : 뉴런과 뉴런 사이에는 전기신호를 통해 정보를 전달한다.
- 인간의 뉴런 구조와 활성화 작동원리를 input(자극) output(반응)과의 연관으로 구현

## 2. 신경 세포 (Neuron)
- 수상돌기 (Dendrite) : 다른 신경세포의 축색돌기와 연결되어 전기화학적 신호를 받아들이는 부위
- 축색돌기 (Axon) : 수신한 전기화학적 신호의 합성결과 값이 특정 임계값이 이상이면 신호를 내보내는 부위
- 신경연접 (Synapse) : 수상돌기와 축색돌기 연결부위, 전달되는 신호의 증폭 또는 감쇄

## 3. 인공 뉴런 (Artificial Neuron)
- 신경세포가 동작하는 원리를 모방을 한것이 인공 뉴런이다.
- 신경세포 구조를 단순화하여 모델링한 구조
- 노드(node)와 엣지(edge)로 표현
- 하나의 노드안에서 입력(input)과 가중치(weights)를 곱하고 더하는 선형구조(linear)
- 활성화 함수(activation function)를 통해서 비선형 구조(non-linear)표현이 가능해지게 된다. > 인공뉴런으로 수학적 모델을 한것

## 4. 인공신경망 구조
- 여러개의 인공뉴런들이 모여 연결된 형태
- 뉴런들이 모인 하나의 다위를 층(layer)이라고 하고, 여러 층(multilayer)으로 이루어 질 수 있다.
- 신경망 계층 예시
  - 입력층(input layer)
  - 은닉층(gidden layer)
  - 출력층(output layer)

인공 신경망은 layer 단위로 순차적으로 쌓이는 구조이지만 실제 신경망은 layer 단위에 구속되지 않고 연결할 수 있다는것이 차이점이다.    
신경세포와 인공뉴런은 여러 신호를 받아 하나의 신호를 흘려주는 동작과 각각이 세포 혹은 노드 단위로 작동하는 구조가 닯아있다.    
또한 신호를 그대로 흘려보내는 것이 아니라 보내지 않을 수도 있고 혹은 다른 크기로 보내는 것이 닮았다.

# 인공신경망 역사

## 1. 최초의 인공신경망
- 인공신경망의경우 약 80년 정도의 역사를 가지고있으며 역사적으로 길지 않다.


- 인간의 신경계를 활성 상태와 비활성 상태를 갖는 이진 뉴런으로 표현 (활성이냐 비활성이냐 두가지 상태로만 표현)
- 생체 뉴런과 같이 시냅스의 흥분과 억제에 따라 신호가 전달되고 늑정 임계치가 넘어야 신호가 발화한다.

## 2. 퍼셉트론 (Perceptron)
**프랭크 로젠블랫(Frank Rosenblatt)은 헵 가설에 따라 인공신경망이 스스로 문제에 맞춰 학습하는 퍼셉트론을 개발했다.**    
최초의 인공신경망이라고 했던 Pitts가 만들었던 모델의 경우 학습과정이 없어 문제에 따라서 신경망의 구성도 매번 변경을 해야하는 문제가 있었다.    
그래서 퍼셉트론이 탄생하게됨    
- 생체신경망을 모방해서 만든 퍼셉트론의 처리 과정과 뉴런의 신호 전달 과정을 비교 해보기

- 하지만 **마빈 민스키(Marvin Lee Minsky)와 시모어 페퍼트(Seymour Papert)는 퍼셉트론이 비선형 문제를 해결할 수 없음을 증명했다** (XOR 논리 연산)
- 이후 퍼셉트론으로 두 개 이상의 직선을 표현하려는 AND 연산이 필요했고 여러 계층으로 표현되는 다층 퍼셉트론(Multi-Layer Perceptron, MLP)탄생했다.

## 3. 다층 퍼셉트론 (Multi-Layer perceptron, MLP)
**여러 개의 퍼셉트론을 층으로 구성한 신경망 모델**    
다층 퍼셉드론을 학습하기 어려운 것은 활성화 함수가 계단 함수라 미분이 불가능하여 미분 가능한 함수가 필요하다.

## 4. 역전파 알고리즘 (Backpropagation)
역전파 알고리즘은 1982년에 폴 워보스 (Paul Werbos)가 학위논문으로 처음 발표하였지만 학위 논문으로만 발표되어 많은 사람에게 알려지지는 않았다.        
이후 1985년에 데이비드 파커(David Parker), 얀 르쿤(Yann Le Cun)이 제안하여 역전파 알고리즘이 이슈가 되었었고,    
1986년에 데이비드 루멜하트(Dacid Rumelhart), 제임스 매클렌런드(James McClelland), 제프리 힌턴(Geoffrey Hinton),    
로널드 윌리엄스(Ronald Williams)가 공개하여 많은 사람들에게 알려지게 되었다.    

이로인해 어떻게 신경망을 계산할 수 있는지 역전파 알고리즘을 통해서 **미분이 가능하고 우리가 최적 회로를 찾을 수 있게 계산이 가능한 알고리즘이 탄생**하게 되었다.

# 딥러닝 역사

## 1. 인공신경망과 딥러닝
일반적으로 인공신경망은 은닉층이 소수인 반면, 딥러닝은 은닉층이 다수 포함된 심층신경망(deep neural network) 형태이다.

딥러닝은 머신러닝의 한 알고리즘으로 인공신경망의 개념과 원리에 뿌리를 두고있다.    
그런데 2000년대 들어서 기존의 인공신경망으로는 한계가 있기 때문에 그 한계를 극복하고     
타 알고리즘의 성능을 뛰어넘었다 이런 의미로써 제프리 힌턴이 딥러닝이라는 새로운 이름을 공개하게 되었다.  

**일반 신경망과 딥러닝 신경망을 비교해서 살펴보기**
1. 일반 신경망
   * 데이터에서 직접 특징(handcrafted feature)을 추출해서 만든 특징 벡터(feature vactor)를 입력으로 사용하고 특징 벡터들의 품질에 영향을준다.
   * 사람들이 직접 특징을 정해주기 때문에 편향이 있을 수 있다.
     
    
2. 딥러닝 신경망
   * 특징추출과 학습을 함께 수행하고 데이터로부터 효과적인 특징을 학습을 통해 추출하여 우수한 성능을 달성한다.
   * 딥러닝은 특징이 중요하면 높은 가중치를 부여하고 중요하지 않으면 낮은 가중치를 부여한다.
   * 그로인해 딥러닝이 스스로 높은 가중치인지 낮은 가중치인지 판단하여 자동으로 특징을 추출하는 기능이 내제되어있다.
   * 딥러닝이 스스로 특징을 추출하기 때문에 편향이 낮다.
   * 추론시간과 성능도 높아진다.
     


    
**딥러닝의 특징**    
딥러닝은 높은 성능과 뛰어난 확장성으로 이미지처리, 자연어 처리, 음성 인식과 같은 분야에서 급격한 발전이 이루어지고 있다.    
딥러닝 모델은 필요한 파라미터(parameter)수가 너무 많아서 더 많은 학습 데이터가 필요하고 최적의 모델과 훈련을 위해 많은 시간과 비용이 필요하다.

## 2. 기울기 소멸 문제 (Vanishing Gradient Problem)
은닉층의 수가 많을수록 더 복잡한 문제를 해결할 수 있지만 너무 깊으면 역전파를 위해    
가중치와 바이어스를 변경하는 수학적 방법인 미분이 불가능해져 0으로 수렴되는, 즉 소멸되는 문제가 있다.
    
* 활성화 함수는 이전단계에서 다음단계에 미치는 영향(가중치)를 찾는데 적합한 통계적 함수
* 효과적인 다층 신경망의 활성화 함수를 찾기 위한 다양한 연구를 제안

**기울기 소멸 문제 해결방안 등장**    
시그모이드나 쌍곡 탄젠트를 사용하지말고 ReLU(Rectified Linear Unit)함수를 사용 하자는 제안을 했다.    
ReLU는 비선형의 구조를 가진다. ReLU를 개선한 ELU가 있다.    
    
기존에는 선형 함수인 시그모이드 함수 등만 적용했으나 비선형 함수인 ReLU,ELU 등 다양한 함수를 적용했더니    
학습도 빠르게 되고 기울기 소멸 문제도 어느정도 해결이 되는 등의 딥러닝 역사가 있었다.

## 3. 과적합 (Overfittiog)
모델이 학습 데이터에 지나치게 맞추어진 상태를 과적합이라고 부르고 과적합된 모델은 학습되지 않는 데이터에 대해 성능 저하를 보인다.    
대부분 모델이 파라미터값이 너무 많아서 과도하게 학습이 되어서 일어나는 현상이다.    
과소적합(underfitting)은 모델이 작거나 단순해서 데이터를 잘 표현하지 못한 상태로 모델의 크기를 키우거나 좋은 모델로 변경이 필요하다.

**과적합을 막기 위해 탄생했던 기법들**
* 규제화 기법 (Regularization) : 큰 가중치 값에 큰 규제를 가하여 과적합되지 않도록 모델을 제한하고 규제의 강도를 정하는 적절한 가중치가 중요하다.
* 드롭아웃 기법 (Dropout) : 노드들을 일정 확률로 무작위 선택해 연결된 가중치 연결선을 없는 것으로 간주하고 학습하는 방법
* 배치 정규화 기법 (Batch Normalization) : 모델에 입력되는 샘플들을 균일하게 만드는 방법으로 미니 배치 단위로 평균이 0, 표준편차가 1이 되도록 정규화하는것

## 4. 딥러닝의 발전
딥러닝 혁명의 아버지, ACM AM Turing Award 수상 제프리 힌튼, 조슈아 벤지오, 얀 르쿤은 인공지능의 주요 혁신을 주도했다.
* 제프리 힌튼
  * 역전파(Backpropagation, 1986)
  * 볼츠만 머신(Boltzmann Machines, 1983)
  * 컨볼루션 신경망 개선(2012)
* 조슈아 벤지오
  * 시퀀스의 확률 모델(1990)
  * 고차원 단어 임베딩 및 어텐션(2000)
  * 생서적 적대 네트워크(2010)
* 얀 르쿤
  * 컨볼루션 신경망(1980)
  * 역전파 알고리즘 개선
  * 신경망의 비전 확장

**현재 많이 사용되는 딥러닝 프레임워크**    
    
텐서플로우 (TensorFlow)    
* 가장 널리 쓰이는 딥러닝 프레임워크 중 하나
* 구글이 주도적으로 개발하는 플랫폼
* 파이썬, C++ API를 기본적으로 제공하고, 자바스크립트(Javascript), 자바(Java), 고(Go), 스위프트(Swift) 다양한 프로그래밍 언어 지원

케라스 (Keras)
* 파이썬으로 작성된 고수준 신경망 API로 TensorFlow, CNTK, 혹은 Theano와 함께 사용 가능
* 사용자 친화성, 모듈성, 확장성을 통해 빠르고 간편한 프로토타이핑 가능

파이토치 (PyTorch)
* 페이스북이 초기 루아(Lua) 언어로 개발된 토치(Torch)를 파이썬 버전으로 개발하여 2017년도에 공개
* 이후 GPU를 이용한 텐서 조작 및 동적 신경망 구축이 가능하도록 딥러닝 프레임워크로 발전시킴
* 파이썬답게 만들어졌고, 유연하면서도 가속화된 계산 속도를 제공

**딥러닝의 많은 기법의 탄생**    
- 전이학습 (transfer learning) : 대용량 데이터로 사전 학습된 모델을 이용해 훈련 시간과 비용을 줄인다.    
- 자기지도학습 (self-supervised learning) : 레이블이 없는 데이터를 자기 스스로 학습    
- 메타 학습 (meta learning) : 학습 방식을 학습 (learn to learn) 하여 인간과 같이 적은 데이터로도 빠르게 학습하는 방법     
- 설명가능한 이공지능 (Explainable, XAI) : 신경망 모델의 내부나 동작 원리를 파악할 수 있도록 하는 기법    
- NAS (Neural Architecture Search) : 최적의 딥러닝 모델을 자동으로 찾아주는 방법    
- AutoML : 강화학습 방식으로 최적의 딥러닝 모델 생성

# 강의정리

## 종합 정리
* 인공신경망 : 인간 두뇌에 대한 계산적 모델을 통해 인공지능을 구현하려는 분야
* 생체신경망을 모방해서 만든 퍼셉트론과 다층퍼셉트론 그리고 역전파 알고리즘이다.
* 인공신경망은 은닉층이 소수인 반면 딥러닝은 은닉층이 다수 포함된 심층신경망이다.
* 딥러닝의 문제점인 기울기 소멸 문제와 과적합 문제를 해결하여 계속해서 발전하고 있다.
